```python
print '---ARRAY BASICS---'
# numpy ndarray
a = np.arange(15)
print a

print '---RESHAPE; SHAPE---'
# make things into an array
a = a.reshape(3, 5)
print a
print a.shape

print '---ROWS and COLUMNS---'
# selecting rows or columns
print a[0,:] # first row (index 0)
print a[0:2,:] # first two rows (python slice style 0:2 => | 0 1 | 2)

# print all the columns
for i in range(a.shape[1]):
    print a[:, i]
    
```

```python
print '---LOGICAL SELECTION---'
AL = [0, 1, 0]
print a[a>5]
print np.where(AL == 1)
print a[np.where(AL == 1), :]

print '---SOME MATH---'
print a - a
print a - 1

print np.log(a+1)
print np.exp(np.log(a+1)) - 1

aa = [0, 1, 1, 1, 0]
bb = [1, 1, 1, 0, 0]

# pairwise iteration with zip
print np.array([x == y for (x, y) in zip(aa, bb)])

print np.array([x == y for (x, y) in zip(aa, bb)]).sum()
```
